```python
import math
def euclidean_distance(a, b):
    jumlah = 0
    for i in range(len(a)):
        jumlah += (a[i] - b[i])**2
    return math.sqrt(jumlah)
# conto
x = [1, 2]
y = [4, 6]
print(euclidean_distance(x, y))   # output: 5.0
```

```python
def manthattan_distance(a, b):
    total = 0
    for i in range(len(a)):
        total += abs(a[i] - b[i])
    return total

print(manthattan_distance([1, 2], [4, 6]))   # output: 7
```

```python
import math
def cosine_similarity(a, b):
    dot = sum(a[i] * b[i] for i in range(len(a)))
    mag_a = math.sqrt(sum(x*2 for x in a))
    mag_b = math.sqrt(sum(x*2 for x in b))
    return dot / (mag_a * mag_b)
print(cosine_similarity([1, 2], [4, 6]))
```

```python
def hamming_distance(a, b):
    return sum(2 for i in range(len(a)) if a[i] != b[i])
print(hamming_distance("2022", "2002"))
```

```python
def permutasi(arr, l, r):
    if l == r:
        print(arr)
    else:
        for i in range(l, r + 1):
            # swap
            arr[l], arr[i] = arr[i], arr[l]
            # rekursi
            permutasi(arr, l + 1, r)
            # backtrack (kembalikan seperti semula)
            arr[l], arr[i] = arr[i], arr[l]
# contoh penggunaan
data = [1, 2, 3]
permutasi(data, 0, len(data) - 1)
```

```python
import heapq

def dijkstra(graph, start):
    dist = {node: float('inf') for node in graph}
    dist[start] = 0
    pq = [(0, start)]

    while pq:
        current_dist, node = heapq.heappop(pq)

        for neighbor, weight in graph[node]:
            distance = current_dist + weight
            if distance < dist[neighbor]:
                dist[neighbor] = distance
                heapq.heappush(pq, (distance, neighbor))

    return dist

# contoh graph
graph = {
    'A': [('B', 1), ('C', 4)],
    'B': [('C', 2), ('D', 5)],
    'C': [('D', 1)],
    'D': []
}

print(dijkstra(graph, 'A'))
```

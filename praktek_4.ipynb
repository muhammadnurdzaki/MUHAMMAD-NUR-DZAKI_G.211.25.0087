```python
def linear_search(arr, target):     #fungsi mencari target dalam array
    for i in range(len(arr)):       #looping dari indeks 0 sampai akhir
        if arr[i] == target:        #cek apakah elemen sama dengan target
            return i                #kalau ketemu kembalikan indeks
    return -1                       #kalau tidak ada,return -1
# contoh
data = [10, 20, 30, 40, 50]
print(linear_search(data, 30)) # output: 2
```

```python
# Binary Search (Iteratif)
def binary_search(arr, target):       #fungsi binary search
    left, right = 0, len(arr) - 1     #batas kiri dan kanan array
    while left <= right:              #selama masih ada range
        mid = (left + right) // 2     #cari indeks tengah
        if arr[mid] == target:        #jika ketemu
            return mid                #kembalikan indeks
        elif arr[mid] < target:       #jika target lebih besar
            left = mid + 1            #geser ke kanan
        else:                         #jika target lebih kecil
            right = mid - 1           #geser ke kiri
    return -1                         #jika tidak di temukan
# contoh (data harus terurut)
data = [10, 20, 30, 40, 50]
print(binary_search(data, 40)) # output: 3
```

```python
# Binary Search (Rekursif)
def binary_search_recursive(arr, left, right, target):                  #fungsi rekursif
    if left > right:                                                    #kondisi berhenti
        return -1                                                       #tidak ditemukan
    mid = (left + right) // 2                                           #cari tengah
    if arr[mid] == target:                                              #jika ketemu
        return mid
    elif arr[mid] < target:                                             #jika target lebih besar
        return binary_search_recursive(arr, mid + 1, right, target)     #ke kanan
    else:                                                               #jika target lebih kecil
        return binary_search_recursive(arr, left, mid - 1, target)      #ke kiri
# contoh
data = [10, 20, 30, 40, 50]
print(binary_search_recursive(data, 0, len(data)-1, 20)) # output: 1
```

```python
# Jump Search
import math                                     #untuk sqrt
def jump_search(arr, target):                   #fungsi jump search
    n = len(arr)                                #panjang array
    step = int(math.sqrt(n))                    #ukuran lompatan
    prev = 0                                    #posisi awal
    # lompat blok
    while arr[min(step, n) - 1] < target:       #lompat per blok
        prev = step                             #simpan posisi sebelumnya
        step += int(math.sqrt(n))               #tambah langkah
        if prev >= n:                           #jika sudah lewat batas
            return -1
    # linear search di blok
    for i in range(prev, min(step, n)):         #cari linier di blok
        if arr[i] == target:                    #jika ketemu
            return i
    return -1                                   #jika tidak ditemukan
# contoh
data = [10, 20, 30, 40, 50, 60, 70]
print(jump_search(data, 60)) # output: 5
```

```python
# Interpolation Search
def interpolation_search(arr, target):                                    #fungsi interpolation search
    low = 0                                                               #indeks awal
    high = len(arr) - 1                                                   #indeks akhir
    while low <= high and target >= arr[low] and target <= arr[high]:     #jika tinggal satu elemen
        if low == high:
            if arr[low] == target:
                return low
            return -1
        # rumus interpolasi
        pos = low + ((target - arr[low]) * (high - low) //                #rumus posisi
                     (arr[high] - arr[low]))
        if arr[pos] == target:                                            #jika ketemu
            return pos
        elif arr[pos] < target:                                           #jika target lebih besar
            low = pos + 1
        else:                                                             #jika target lebih kecil
            high = pos - 1
    return -1                                                             #jika tidak ditemukan
# contoh
data = [10, 20, 30, 40, 50]
print(interpolation_search(data, 50)) # output: 4
```

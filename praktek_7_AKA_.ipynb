```python
# Program PCA Sederhana
data = [[1, 2],
        [2, 1],
        [3, 5],
        [4, 4]]

mean = []
for kolom in range(len(data[0])):
    jumlah = 0
    for baris in range(len(data)):
        jumlah += data[baris][kolom]
    mean.append(jumlah / len(data))
print("Mean Data:")
print(mean)
center = []
for baris in range(len(data)):
    temp = []
    for kolom in range(len(data[baris])):
        nilai = data[baris][kolom] - mean[kolom]
        temp.append(nilai)
    center.append(temp)
print("\nData Centering:")
for i in center:
    print(i)
transpose = []
for kolom in range(len(center[0])):
    temp = []
    for baris in range(len(center)):
        temp.append(center[baris][kolom])
    transpose.append(temp)
cov_matrix = []
for i in range(len(transpose)):
    baris_cov = []
    for j in range(len(center[0])):
        total = 0
        for k in range(len(center)):
            total += transpose[i][k] * center[k][j]
        hasil_cov = total / (len(center) - 1)
        baris_cov.append(hasil_cov)
    cov_matrix.append(baris_cov)
print("\nCovariance Matrix:")
for i in cov_matrix:
    print(i)
```

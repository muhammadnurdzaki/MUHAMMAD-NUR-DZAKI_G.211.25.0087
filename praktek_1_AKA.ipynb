```python
def total(angka ,n):
  jumlah = 0
  for i in range (n):
    jumlah = jumlah + angka [i]
  return jumlah
angka = [4,5,6,3]
n = len (angka)
print(total (angka,n))
```

```python
a= 3
b= 3
angka =[(4,2,6), (3,7,9), (2,5,8)]
total = 0

for i in range(a):
  for j in range(b):
    total += angka [i][j]
print (total)
```
